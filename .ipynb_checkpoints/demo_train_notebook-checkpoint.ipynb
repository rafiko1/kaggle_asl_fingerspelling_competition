```python
with strategy.scope():
    model = get_model(CFG, dtype=tf.bfloat16)
    
    # Optimizer (w/wo schedule)
    opt = get_optimizer(CFG)
    
    # Callbacks
    callbacks = get_callbacks(CFG, model)
    
    # Loss
    loss = get_loss(CFG)
    
    model.compile(
        optimizer=opt,
        loss = loss
        # loss = SCCE_loss
    )
```

```python
history = model.fit(
    train_ds,
    epochs=CFG['n_epochs'],
    verbose=CFG['verbose'],
    validation_data=valid_ds,
    callbacks = callbacks
)
```
